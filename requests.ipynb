```python
nome_artistas = [] 
first = 0
last = 5

for track in tracks[first:last]:
    artists = track['track']['album']['artists']  # Obtenha a lista de artistas da faixa
    for artist in artists:
        artist_name = artist['name']  # Acesse o nome de cada artista
        nome_artistas.append(artist_name)

# Imprimir os nomes de todos os artistas
for artist in nome_artistas:
    print(f'Nome do Artista: {artist}')
```
